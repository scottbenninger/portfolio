```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url = "https://raw.githubusercontent.com/dsc-courses/dsc106-wi24/gh-pages/resources/data/weather.csv"
data = pd.read_csv(url)

# Aggregate data by city and sort by rainfall
avg_rainfall = data.groupby('city')['rain'].mean().reset_index()
avg_rainfall = avg_rainfall.sort_values(by='rain', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=avg_rainfall, x='rain', y='city', palette='RdYl
Bu_r')

# Add value annotations on bars
for i, v in enumerate(avg_rainfall['rain']):
    ax.text(v + 0.1, i, f"{v:.2f}", color='black', va='center', fontsize=10)

# Add title, labels, and grid
plt.title("Average Monthly Rainfall by City", fontsize=16)
plt.xlabel("Rainfall (inches)", fontsize=12)
plt.ylabel("City", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Improve layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
url = "https://raw.githubusercontent.com/dsc-courses/dsc106-wi24/gh-pages/resources/data/weather.csv"
data = pd.read_csv(url)

# Aggregate data: Average rainfall and sunshine for each city
city_data = data.groupby(['city', 'lat', 'lon']).agg(
    avg_rainfall=('rain', 'mean'),
    avg_sunshine=('sunshine', 'mean')
).reset_index()

# Plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    city_data['lon'], 
    city_data['lat'], 
    c=city_data['avg_rainfall'], 
    s=city_data['avg_sunshine'] / 2,  # Scale bubble size
    cmap='Blues', 
    edgecolor='black', 
    alpha=0.8
)

# Add city labels
for _, row in city_data.iterrows():
    plt.text(
        row['lon'] + 0.5, row['lat'], row['city'], 
        fontsize=10, ha='left', va='center'
    )

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label("Average Rainfall (inches)", fontsize=12)

# Title and axis labels
plt.title("Climate Patterns: Rainfall and Sunshine Across Major U.S. Cities", fontsize=14)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)

# Grid for clarity
plt.grid(True, linestyle='--', alpha=0.5)

# Annotate legend for bubble size
plt.text(
    -130, 50, "Bubble Size = Average Sunshine Hours", 
    fontsize=10, color='gray', style='italic'
)

# Final adjustments
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url = "https://raw.githubusercontent.com/dsc-courses/dsc106-wi24/gh-pages/resources/data/weather.csv"
data = pd.read_csv(url)

# Find extreme months for rainfall and sunshine for each city
extremes = data.groupby('city').agg(
    max_rain=('rain', 'max'),
    min_rain=('rain', 'min'),
    max_sunshine=('sunshine', 'max'),
    min_sunshine=('sunshine', 'min')
).reset_index()

# Normalize data for better comparison
extremes['rain_range'] = extremes['max_rain'] - extremes['min_rain']
extremes['sunshine_range'] = extremes['max_sunshine'] - extremes['min_sunshine']

# Plot side-by-side bars for rain and sunshine ranges
plt.figure(figsize=(12, 6))
bar_width = 0.4
cities = extremes['city']
x = range(len(cities))

# Rainfall range
plt.bar(x, extremes['rain_range'], width=bar_width, label='Rainfall Range (inches)', color='skyblue', edgecolor='black')

# Sunshine range
plt.bar([p + bar_width for p in x], extremes['sunshine_range'], width=bar_width, label='Sunshine Range (hours)', color='orange', edgecolor='black')

# Labels and title
plt.title('Climate Extremes: Rainfall and Sunshine Variability Across U.S. Cities', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Range (Max - Min)', fontsize=12)
plt.xticks([p + bar_width / 2 for p in x], cities, rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
url = "https://raw.githubusercontent.com/dsc-courses/dsc106-wi24/gh-pages/resources/data/weather.csv"
data = pd.read_csv(url)

# Calculate averages for each city
city_data = data.groupby('city').agg(
    avg_rain=('rain', 'mean'),
    avg_sunshine=('sunshine', 'mean')
).reset_index()

# Plot settings
bar_width = 0.4
x = np.arange(len(city_data['city']))

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x, city_data['avg_rain'], width=bar_width, label='Average Rainfall (inches)', color='skyblue', edgecolor='black')
plt.bar(x + bar_width, city_data['avg_sunshine'], width=bar_width, label='Average Sunshine (hours)', color='orange', edgecolor='black')

# Add labels and title
plt.title('Average Rainfall and Sunshine Across U.S. Cities', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Value', fontsize=12)
plt.xticks(x + bar_width / 2, city_data['city'], rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
